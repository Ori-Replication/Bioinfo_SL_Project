```python
import pandas as pd
```

```python
primekg_path = '../data/download_data/PrimKG.csv'
primekg_df=pd.read_csv(primekg_path)
```

```python
x_gene_protein_df = primekg_df[primekg_df['x_type'] == 'gene/protein' ]
y_gene_protein_df = primekg_df[primekg_df['y_type'] == 'gene/protein' ]
```

```python

x_gene_protein = x_gene_protein_df['x_name'].values
y_gene_protein = y_gene_protein_df['y_name'].values
gene_set = set(x_gene_protein).union(set(y_gene_protein))
print(len(gene_set))
```

```python
gene_set
```

```python
# Give it a id and create a dict
gene_dict = {gene: idx for idx, gene in enumerate(gene_set)}
gene_dict
```

```python
out_json = {'gene/protein': gene_dict}
out_json
```

```python
import json

json.dump(out_json, open('../data/processed_data/gene_protein_2_id.json', 'w'))
```
