```python
import sys
import os
import icecream as ic

ori_data_dir = '../data/original_data'
A375_data = pd.read_csv(os.path.join(ori_data_dir, 'A375.csv'))
A549_data = pd.read_csv(os.path.join(ori_data_dir, 'A549.csv'))
Jurkat_data = pd.read_csv(os.path.join(ori_data_dir, 'Jurkat.csv'))
```

```python
def get_unique_gene_list(data):
    gene_list = []
    for i in range(data.shape[0]):
        gene_1 = data['gene1_name'][i]
        gene_2 = data['gene2_name'][i]
        if gene_1 not in gene_list:
            gene_list.append(gene_1)
        if gene_2 not in gene_list:
            gene_list.append(gene_2)
    return gene_list

A375_gene_list = get_unique_gene_list(A375_data)
A549_gene_list = get_unique_gene_list(A549_data)
Jurkat_gene_list = get_unique_gene_list(Jurkat_data)
```

```python
# 1. get the gene index (column name)
CN_df_gene_index = CN_df.columns
CN_df_gene_index
gene_2_index = {}
for gene in CN_df_gene_index:
    # split the gene name according to '(' ')'
    gene_name = gene.split('(')[0]
    gene_name = gene_name.strip()
    gene_index = gene.split('(')[1]
    # if not number
    if not gene_index[0].isdigit():
        print('not number' + gene_index)
        
    gene_index = gene_index.split(')')[0]
    gene_index = gene_index.strip()
    gene_2_index[gene_name] = gene_index

print(len(gene_2_index))
```

```python
# generate the gene indes in the form of 'gene_name (index)'
A375_gene_index_list = []
for gene in A375_gene_list:
    if gene in gene_2_index:
        index = gene_2_index[gene]
        A375_gene_index_list.append(gene + ' (' + index + ')')
    else:
        A375_gene_index_list.append(gene + ' (NA)')
    
A549_gene_index_list = []
for gene in A549_gene_list:
    if gene in gene_2_index:
        index = gene_2_index[gene]
        A549_gene_index_list.append(gene + ' (' + index + ')')
    else:
        A549_gene_index_list.append(gene + ' (NA)')

Jurkat_gene_index_list = []
for gene in Jurkat_gene_list:
    if gene in gene_2_index:
        index = gene_2_index[gene]
        Jurkat_gene_index_list.append(gene + ' (' + index + ')')
    else:
        Jurkat_gene_index_list.append(gene + ' (NA)')


```

```python
# find if the key exists in the dictionary, for the gene in the A549_gene_list
for gene in A549_gene_list:
    if gene not in gene_2_index:
        print(gene)

# find if the key exists in the dictionary, for the gene in the Jurkat_gene_list
for gene in Jurkat_gene_list:
    if gene not in gene_2_index:
        print(gene)
```

```python
# filter out the gene in the A375_gene_list 
# first, map to index
# then, use the index to filter out the gene.use 


CN_df_A375_gene = pd.DataFrame()
for gene_index_ in A375_gene_index_list:
    if gene_index_ in CN_df_A375.columns:
        CN_df_A375_gene[gene_index_] = CN_df_A375[gene_index_]
        
    else:
        CN_df_A375_gene[gene_index_] = np.nan
        
CN_df_A549_gene = pd.DataFrame()
for gene_index_ in A549_gene_index_list:
    if gene_index_ in CN_df_A549.columns:
        CN_df_A549_gene[gene_index_] = CN_df_A549[gene_index_]
    else:
        CN_df_A549_gene[gene_index_] = np.nan
        
CN_df_Jurkat_gene = pd.DataFrame()
for gene_index_ in Jurkat_gene_index_list:
    if gene_index_ in CN_df_JURKAT.columns:
        CN_df_Jurkat_gene[gene_index_] = CN_df_JURKAT[gene_index_]
    else:
        CN_df_Jurkat_gene[gene_index_] = np.nan
```
