```python
import pandas as pd
```

```python
data_root_path = '../data/processed_data/'
A375_CN_data = pd.read_csv(data_root_path + 'CN_A375_gene.csv', index_col=0)
A549_CN_data = pd.read_csv(data_root_path + 'CN_A549_gene.csv', index_col=0)
Jurkat_CN_data = pd.read_csv(data_root_path + 'CN_Jurkat_gene.csv', index_col=0)
A375_Expression_data = pd.read_csv(data_root_path + 'Expression_A375_gene.csv', index_col=0)
A549_Expression_data = pd.read_csv(data_root_path + 'Expression_A549_gene.csv', index_col=0)
Jurkat_Expression_data = pd.read_csv(data_root_path + 'Expression_Jurkat_gene.csv', index_col=0)
A375_HotspotMutation_data = pd.read_csv(data_root_path + 'HotspotMutation_A375_gene.csv', index_col=0)
A549_HotspotMutation_data = pd.read_csv(data_root_path + 'HotspotMutation_A549_gene.csv', index_col=0)
Jurkat_HotspotMutation_data = pd.read_csv(data_root_path + 'HotspotMutation_Jurkat_gene.csv', index_col=0)
```

```python
# 将数据进行合并，行标识为基因名，列标识为数据种类
A375_data = pd.concat([A375_CN_data, A375_Expression_data, A375_HotspotMutation_data], axis=0)
A375_data = A375_data.T
A375_data.columns = ['CN', 'Expression', 'HotspotMutation']
A375_data.index = A375_data.index.str.split('(').str[0]
# 将现在的index变为第一列
A375_data.reset_index(inplace=True)
# 将第一列的列名改为 Gene Name
A375_data.rename(columns={'index': 'Gene Name'}, inplace=True)
A375_data['Gene Name'] = A375_data['Gene Name'].str.strip()
A375_data.fillna(0, inplace=True)
```

```python
A549_data = pd.concat([A549_CN_data, A549_Expression_data, A549_HotspotMutation_data], axis=0)
A549_data = A549_data.T
A549_data.columns = ['CN', 'Expression', 'HotspotMutation']
A549_data.index = A549_data.index.str.split('(').str[0]
A549_data.reset_index(inplace=True)
A549_data.rename(columns={'index': 'Gene Name'}, inplace=True)
A549_data['Gene Name'] = A549_data['Gene Name'].str.strip()
A549_data.fillna(0, inplace=True)
```

```python
Jurkat_data = pd.concat([Jurkat_CN_data, Jurkat_Expression_data, Jurkat_HotspotMutation_data], axis=0)
Jurkat_data = Jurkat_data.T
Jurkat_data.columns = ['CN', 'Expression', 'HotspotMutation']
Jurkat_data.index = Jurkat_data.index.str.split('(').str[0]
Jurkat_data.reset_index(inplace=True)
Jurkat_data.rename(columns={'index': 'Gene Name'}, inplace=True)
# strip gene name
Jurkat_data['Gene Name'] = Jurkat_data['Gene Name'].str.strip()
Jurkat_data.fillna(0, inplace=True)
```

```python
# save to csv
A375_data.to_csv(data_root_path + 'A375_all_data_gene.csv', index=False)
A549_data.to_csv(data_root_path + 'A549_all_data_gene.csv', index=False)
Jurkat_data.to_csv(data_root_path + 'Jurkat_all_data_gene.csv', index=False)
```
