```python
import pandas as pd
import numpy as np
```

```python
Jurkat_path = '../data/train_data/Jurkat_filtered.csv'

Jurkat_df = pd.read_csv(Jurkat_path)
```

```python
Jurkat_df
```

```python
positive_samples = Jurkat_df[Jurkat_df['label'] == 1]
negative_samples = Jurkat_df[Jurkat_df['label'] == 0]

# Apply the 1/9 probability retention to negative samples
negative_samples_resampled = negative_samples.sample(frac=1/9, random_state=42)

# Concatenate the positive samples with the resampled negative samples
resampled_data = pd.concat([positive_samples, negative_samples_resampled])

# shuffle
resampled_data = resampled_data.sample(frac=1, random_state=42)

```

```python
resampled_data
```

```python
# reindex
resampled_data = resampled_data.reset_index(drop=True)

resampled_data
```

```python
# calculate the positive rate after the resample
positive_rate = resampled_data['label'].sum() / len(resampled_data)
positive_rate
```

```python
# good. save to csv
resampled_data.to_csv('../data/train_data/Jurkat_resampled.csv', index=False)
```
