```python
entity_set_x = set(primkg_df['x_name'])
entity_set_y = set(primkg_df['y_name'])
entity_set = entity_set_x.union(entity_set_y)
```

```python
gene_A375_data_path = '../data/original_data/A375.csv'
gene_A549_data_path = '../data/original_data/A549.csv'
gene_Jurkat_data_path = '../data/original_data/Jurkat.csv'
gene_A375_df = pd.read_csv(gene_A375_data_path)
gene_A549_df = pd.read_csv(gene_A549_data_path)
gene_Jurkat_df = pd.read_csv(gene_Jurkat_data_path)
# gene_set = set(gene_A375_df['gene1_name'] ,gene_A375_df['gene2_name'] , gene_A549_df['gene1_name'] , gene_A549_df['gene2_name'] , gene_Jurkat_df['gene1_name'] , gene_Jurkat_df['gene2_name'])
gene_set = set(gene_A375_df['gene1_name']).union(set(gene_A375_df['gene2_name'])).union(set(gene_A549_df['gene1_name'])).union(set(gene_A549_df['gene2_name'])).union(set(gene_Jurkat_df['gene1_name'])).union(set(gene_Jurkat_df['gene2_name']))
```

```python
# entity 2 id
entity2id = {}
iter = 0
for entity in gene_set:
    entity2id[entity] = primkg_df[primkg_df['x_name']== entity].iloc[0]['x_id']
    # if do not exist in x_name
    if pd.isnull(entity2id[entity]):
        print(entity, 'not in x_name')
        break
    print(entity, entity2id[entity])
    iter += 1
    print(iter)
```
